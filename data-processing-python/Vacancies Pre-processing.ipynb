```python
import pandas as pd
import numpy  as np

# Read all datasets as each represents a month in 2016
vacancies_1_df = pd.read_csv("jad8jan2016.csv")
vacancies_2_df = pd.read_csv("jad8feb2016.csv")
vacancies_3_df = pd.read_csv("jad8mac2016.csv")
vacancies_4_df = pd.read_csv("jad8apr2016.csv")
vacancies_5_df = pd.read_csv("jad8mei2016.csv")
vacancies_6_df = pd.read_csv("jad8jun2016.csv")
vacancies_7_df = pd.read_csv("jad8jul2016.csv")
vacancies_8_df = pd.read_csv("jad8aug2016.csv")
vacancies_9_df = pd.read_csv("jad8sept2016.csv")
vacancies_10_df = pd.read_csv("jad8okt2016.csv")
vacancies_11_df = pd.read_csv("jad8nov2016.csv")
vacancies_12_df = pd.read_csv("jad8dis2016.csv")

# Combine all the dataframe read into a list so easier
# manipulation
vacancies_df_list = [vacancies_1_df,
                     vacancies_2_df,
                     vacancies_3_df,
                     vacancies_4_df,
                     vacancies_5_df,
                     vacancies_6_df,
                     vacancies_7_df,
                     vacancies_8_df,
                     vacancies_9_df,
                     vacancies_10_df,
                     vacancies_11_df,
                     vacancies_12_df]
```

```python
# Manually insert the title for the dataframe for easier manipulation
# *All the titles are manually extracted & slightly modified from the dataset itself*
# There must be a better way to handle this but this way is quicker.
new_title = ["State",
             "Agriculture, Forestry & Fishing",
             "Mining & Quarrying",
             "Manufacturing",
             "Electricity, Gas, Steam & Air Conditioning Supply",
             "Water Supply, Sewerage, Waste Management & Remediation Activities",
             "Construction",
             "Wholesale & Retail Trade, Repair of Motor Vehicles & Motorcycles",
             "Transportation & Storage",
             "Accomodation & Food Service Activities",
             "Information & Communication",
             "Financial & Insurance / Takaful Activities",
             "Real Estate Activities",
             "Professional, Scientific & Technical Activities",
             "Administrative & Support Service Activities",
             "Public Administration & Defence",
             "Education",
             "Human Health & Social Work Activities",
             "Arts, Entertainment & Recreational",
             "Other Service Activities",
             "Activities of Households as Employers",
             "Activities of Extraterritorial Organisation & Bodies",
             "Total"]
```

```python
# Randomly read a csv file so that the new dataframe has a structure
# (number of rows & columns)
vacancies_df = pd.read_csv("jad8jan2016.csv")

# Drop any row that contains NaN value
vacancies_df = vacancies_df.dropna()

# Reset the columns' titles
vacancies_df.columns = new_title

# Get all the indices for easier manipulation later on
index_list = vacancies_df.index.tolist()

# Reset the index of the dataframe to be the state instead
vacancies_df.set_index("State", inplace=True)

# Reset all the values in the dataframe to 0
for title in new_title[1:]:
    vacancies_df[title] = 0

# Show first 5 rows of the newly prepared dataframe
vacancies_df.head()    
```

```python
# Sum together each cell value from 12 datasets and fill them into
# the newly prepared dataframe

for title in new_title[1:]:
    values_list = [0] * len(index_list)
    for df in vacancies_df_list:
        df = df.dropna()
        df.columns = new_title
        i = 0
        for index in index_list:
            values_list[i] += float(df.loc[index, title].replace(',',''))
            i += 1
    vacancies_df[title] = values_list
       
```

```python
# Make a new copy of the dataframe
preprocessed_vac_df = vacancies_df.copy()

# Drop all the unnecessary columns
state_to_drop = [' - Labuan',
                 ' - Putrajaya', 
                 'Jumlah Kecil-S.Malaysia', 
                 'Jumlah Kecil -(Sub -Total)', 
                 'Jumlah Besar']

preprocessed_vac_df.drop(state_to_drop, inplace=True)

# Rename the index to make them consistent
index_lst = preprocessed_vac_df.index.tolist()
index_list = [i.replace(' - ', 'W.P. ') for i in index_lst]

preprocessed_vac_df.index = index_list

# Sort the dataframe by their index
preprocessed_vac_df = preprocessed_vac_df.sort_index()

# Show the preprocessed dataframe
preprocessed_vac_df
```
